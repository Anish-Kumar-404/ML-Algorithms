```python
import mlxtend
import pandas as pd
import numpy as np
df=pd.read_csv(r"wine.csv")
df.head()
```

```python
df1=df.pop('Wine')
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, df1, test_size = 0.2)

#from sklearn.ensemble import RandomForestClassifier
#rfc=RandomForestClassifier()

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#cv =  cross validation
from mlxtend.feature_selection import SequentialFeatureSelector
sfs1=SequentialFeatureSelector(lr, k_features=(1,13), forward=False, verbose=4, cv=5)
sfs1.fit(x_train, y_train)
```

```python
pd.DataFrame.from_dict(sfs1.get_metric_dict()).T #.T is for transform
```
